### Importamos y cargamos

```python
import pandas as pd
import altair as alt
from google.colab import files 
import io 
import requests
```

### Vistas previas

```python
url = "https://raw.githubusercontent.com/riclops17/DatosVis/main/resultados-definitivos-elecciones-generales-2019.csv"
s= requests.get(url).content
df2019= pd.read_csv(io.StringIO(s.decode('utf-8')))
```

```python

df2019.head()
```

```python
df2021 = pd.read_csv('https://aaizemberg.github.io/2021/data/mesas.zip', sep='\t', dtype= {
    'Agrupacion' : object,
    'Codigo': object,  
    'IdCargo': object,  
    'IdDistrito': object,  
    'IdSeccion': object,  
    'IdAgrupacion': object,  
    'IdCircuito' : object})
df2021.head()
```

```python
data2019 = df2019.groupby(by=["cargo_descripcion", "partido_descripcion","comuna"])['total_votos'].agg('sum').to_frame().reset_index()
data2019.head()
```

## **Visualización 1**

Por comuna y agrupación/partido descripción cuantos votos se realizaron, se filtra tambien 3 tipos de votos: votos en blanco, nulos y recurridos en las elecciones del 2019: 

* Votos en blanco: Se manifiesta que ninguna de las opciones planteadas es adecuada. 
* Voto nulo: Es un voto mal realizado en una elección, al punto que acarrea su nulidad
* Voto recurrido: Es un voto que no se computa. Será analizado y clasificado en el escrutinio definitivo, momento en el que será volcado en otras clases de votos.

```python
alt.Chart(data2019).mark_bar().encode(
    x='sum(total_votos)',
    #y='comuna',
    y=alt.Y('comuna', sort='x'),
    #color='partido_descripcion:N',
    color = alt.Color('partido_descripcion:N', scale=alt.Scale(domain = ["ALIANZA CONSENSO FEDERAL",
          "ALIANZA FRENTE DE TODOS","ALIANZA JUNTOS POR EL CAMBIO",
          "FRENTE IZQ Y DE TRAB-UNIDAD",
          "UNITE POR LA LIBERTAD Y LA DIG","VOTOS DE IDENTIDAD IMPUGNADOS","VOTOS EN BLANCO",
          "VOTOS NULOS","VOTOS RECURRIDOS"], range=["#152f8f","#1f95db", "   #fcc404", "  #e65f80",
          " #084ca5",'grey','grey',"grey","grey"])),
    order=alt.Order(
      # Sort the segments of the bars by this field
      'partido_descripcion',
      sort='ascending'
    )
    
).properties(width=400,height=300,title="Elecciones 2019")
```

## **Visualización 2**

Por comuna y por agrupación/partido descripción vemos la cantidad de votos que se realizaron en las paso de 2021 diferenciando en diputados nacionales y provinciales.

```python
data2021 = df2021.groupby(by=["Cargo", "Agrupacion","Seccion"])['votos'].agg('sum').to_frame().reset_index()
```

```python
alt.Chart(data2021).mark_bar().encode(
    x='sum(votos)',
    y=alt.Y('Seccion:N', sort='x'),
    #y='Seccion',
    color= alt.Color('Agrupacion:N', scale=alt.Scale(domain = ["JUNTOS POR EL CAMBIO",
          "FRENTE DE TODOS", "LA LIBERTAD AVANZA",
          "FRENTE DE IZQUIERDA Y DE TRABAJADORES - UNIDAD",
          "AUTODETERMINACIÓN Y LIBERTAD", "AUTONOMISTA", "FEDERAL",
          "MOVIMIENTO AL SOCIALISMO", "MOVIMIENTO LIBRES DEL SUR",
          "POLÍTICA OBRERA", "SOCIALISTA AUTÉNTICO",
          "DIGNIDAD POPULAR", "APTITUD RENOVADORA",
          "RENOVADOR FEDERAL"], range=["gold", "blue", "magenta", "red",
          "#592E6B",'purple','pink','black','lightblue','red','red','green','grey','brown'])),

    order=alt.Order(
      # Sort the segments of the bars by this field
      'Agrupacion',
      sort='ascending'
    )
).facet(
    column='Cargo:N'
)
```
